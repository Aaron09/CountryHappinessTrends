```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
data_2016 = pd.read_csv('2016.csv')
```

```python
countries, gdp = (data_2016['Country'].tolist(), data_2016['Economy (GDP per Capita)'].tolist())
```

```python
# GDP FIGURE

x_pos = np.arange(len(countries))
plt.plot(x_pos, gdp, 'bo')
plt.ylabel("GDP per Capita")
plt.xlabel("Most happy countries -> Least happy countries")
plt.show()
```

```python
countries, life_expectancies = (data_2016['Country'].tolist(), data_2016['Health (Life Expectancy)'].tolist())
```

```python
# LIFE EXPECTANCY FIGURE

x_pos = np.arange(len(countries))
plt.plot(x_pos, life_expectancies, 'go')
plt.ylabel("Life Expectancy")
plt.xlabel("Most happy countries -> Least happy countries")
plt.show()
```

```python
countries, trust_values = (data_2016['Country'].tolist(), data_2016['Trust (Government Corruption)'].tolist())
```

```python
# GOV TRUST FIGURE

x_pos = np.arange(len(countries))
plt.plot(x_pos, trust_values, 'ro')
plt.ylabel("Level of Government trust")
plt.xlabel("Most happy countries -> Least happy countries")
plt.show()
```

```python
terrorism_df = pd.read_csv('terrorism_data.csv', encoding='latin1', low_memory=False)
```

```python
terror_2016 = terrorism_df.loc[terrorism_df['iyear'] == 2016]
terror_counts = terror_2016['country_txt'].value_counts()
terror_dict = dict(zip(terror_counts.keys().tolist(), terror_counts.tolist()))

# some countries were not present in the terrorism dataset, so we avoid those.
# I am not sure why they are not in the dataset, potentially because they did not have a terrorist attack.
valid_countries = []
terror_in_happy_order = []
for country in countries:
    if country in terror_dict:
        terror_in_happy_order.append(terror_dict[country])
        valid_countries.append(country)
```

```python
# TERRORISM FIGURE

x_pos = np.arange(len(valid_countries))
plt.plot(x_pos, terror_in_happy_order, 'ro')
plt.ylabel("Terrorism Incidents in 2016")
plt.xlabel("Most happy countries -> Least happy countries")
plt.show()
```
